```python
#drop
df=df.drop(['Running Nose', 'Chronic Lung Disease', 'Headache', 'Heart Disease', 'Diabetes', 'Gastrointestinal ', 'Wearing Masks', 'Sanitization from Market', 'Asthma', 'Fatigue '],axis=1)
```

```python
x=df.drop('COVID-19',axis=1)
#get_col
y=df['COVID-19']
```

```python
#train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state=42)
```

```python
class LogisticRegression(torch.nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LogisticRegression, self).__init__()
        self.linear = torch.nn.Linear(input_dim, output_dim)
        
    def forward(self, x):
        outputs = torch.sigmoid(self.linear(x))
        return outputs
```

```python
#shape
input_dim = x_train.shape[1]
output_dim = 1 
#init_model
model = LogisticRegression(input_dim,output_dim)
loss_fn = torch.nn.BCELoss()
inputs = torch.Tensor(x_train.values)
outputs = torch.Tensor(np.transpose([y_train.values]))
opt = torch.optim.SGD(model.parameters(), lr=0.5)
```

```python
#get_grad
#grad_agg
#grad_update
for epoch in range(100):
    if(epoch%10==0):
          print("processing round: "+str(epoch))
    pred = model(inputs)
    #print(pred)
    loss = loss_fn(pred, outputs)
    loss.backward()
    
    gradients = []
    for p in model.parameters():
        #print(p.grad)
        gradients.append(p.grad)
    #print("----------grads----------")
    #print(gradients)
    for i, p in enumerate(model.parameters()):
        p.grad = copy.deepcopy(gradients[i])
        print(p.grad)
    print("=============grads============")
    for p in opt.param_groups[0]['params']:
        if p.grad is None:
            print("p is none")
        else:
            print(p.grad)
    #print("----------before----------")
    #print(model.state_dict())
    #print(opt.param_groups)
    opt.step()
    opt.zero_grad()
    #print("----------after----------")
    #print(model.state_dict())
    #print(opt.param_groups)
```
